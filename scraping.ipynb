```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL of the webpage you want to scrape
url = 'https://euforbih.org/index.php/bih-minefield-maps'


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    'Referer': 'https://www.google.com'
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page with Beautiful Soup
    soup = BeautifulSoup(response.text, 'lxml')
    
    # Extract elements from the BeautifulSoup object
    # For example, to find all <a> tags with 'href' attribute
    links = soup.find_all('area', href=True)
    
    # Print the href attributes of the found tags
    for link in links:
        print(link['href'])
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import time
import random


for link in links:
    url = "https://euforbih.org/" + link['href']

    response = requests.get(url, headers=headers)

    id_pdf = url.split("/")[-1]
    # Check if the request was successful
    if response.status_code == 200:
        # Write the content of the response to a file
        with open(f'data/{id_pdf}', 'wb') as file:
            file.write(response.content)
        print(f"The PDF file {id_pdf} has been downloaded successfully.")
    else:
        print(f"Failed to download the PDF {id_pdf}. Status code: {response.status_code}")
    
    sleep_time = random.uniform(1, 2)  # Generates a random float between 1 and 2
    print(f"Pausing for {sleep_time:.2f} seconds.")
    time.sleep(sleep_time)  # Pauses execution for the sleep_time duration
```
